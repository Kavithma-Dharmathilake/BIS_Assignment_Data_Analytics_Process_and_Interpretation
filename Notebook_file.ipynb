## Add Colab Notebook to GitHub

To push this notebook to your GitHub repository, we will follow these steps:

1.  **Initialize Git Repository:** If you haven't already, initialize a Git repository in your Colab environment.
2.  **Configure Git User:** Set your Git username and email.
3.  **Add Remote Repository:** Connect your local repository to your GitHub repository.
4.  **Stage Notebook File:** Add the current notebook file to the Git staging area.
5.  **Commit Changes:** Record the staged changes with a descriptive message.
6.  **Push to GitHub:** Upload your local commits to your remote GitHub repository.

### Step 1: Initialize Git Repository

Run the following command to initialize a new Git repository in your current directory. If you've already done this successfully, you can skip this cell.

```python
!git init
```

```python
# Replace with your GitHub repository HTTPS URL
!git remote add origin https://github.com/Kavithma-Dharmathilake/BIS_Assignment_Data_Analytics_Process_and_Interpretation.git
```

### Step 4: Stage the Notebook File

Add the current notebook file to the staging area. Colab notebooks are typically saved with a `.ipynb` extension. You can find the exact name of your current notebook from the Colab interface (usually at the top of the page).
```python
!git commit -m "Initial Commit from Colab"
```

Finally, push your committed changes to your GitHub repository. The first time you push, you'll typically use `-u origin main` (or `master` if that's your default branch name) to set the upstream tracking branch.

```python
# Use 'main' or 'master' depending on your repository's default branch name
!git push -u origin main
```
