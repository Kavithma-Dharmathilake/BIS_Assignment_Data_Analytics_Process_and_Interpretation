The dataset represents ten years (1999-2008) of clinical care at 130 US hospitals and integrated delivery networks. Each row concerns hospital records of patients diagnosed with diabetes, who underwent laboratory, medications, and stayed up to 14 days. The goal is to determine the early readmission of the patient within 30 days of discharge. The problem is important for the following reasons. Despite high-quality evidence showing improved clinical outcomes for diabetic patients who receive various preventive and therapeutic interventions, many patients do not receive them. This can be partially attributed to arbitrary diabetes management in hospital environments, which fail to attend to glycemic control. Failure to provide proper diabetes care not only increases the managing costs for the hospitals (as the patients are readmitted) but also impacts the morbidity and mortality of the patients, who may face complications associated with diabetes.

### Encode Variables

Under this section categorical variables will be encoded.

The values of age are given like [0–10),[10–20),[20–30)

The age feature is ordinal and it is converted as bellow,

[0–10) → 5, [10–20) → 15, [20–30) — -> 25

```python
#Encode Age Values

import re

def convert_age(age_range):
    if pd.isna(age_range):
        return age_range
    numbers = list(map(int, re.findall(r'\d+', age_range)))
    return sum(numbers) / 2

df['age'] = df['age'].apply(convert_age)
```

```python
# Function to Map ICD Codes

def map_icd9_to_category(code):
    if pd.isna(code):
        return 'Other'

    # Handle E codes
    if code.startswith('E'):
        return 'External Causes / Injury'

    # Handle V codes
    if code.startswith('V'):
        return 'Supplemental / Other'

    # Convert numeric part to int
    try:
        num = int(code[:3])
    except:
        return 'Other'

    if 1 <= num <= 139:
        return 'Infectious and parasitic diseases'
    elif 140 <= num <= 239:
        return 'Neoplasms'
    elif 240 <= num <= 279:
        return 'Endocrine, nutritional, metabolic, immunity disorders'
    elif 280 <= num <= 289:
        return 'Blood diseases'
    elif 290 <= num <= 319:
        return 'Mental disorders'
    elif 320 <= num <= 389:
        return 'Nervous system & sense organs'
    elif 390 <= num <= 459:
        return 'Circulatory system'
    elif 460 <= num <= 519:
        return 'Respiratory system'
    elif 520 <= num <= 579:
        return 'Digestive system'
    elif 580 <= num <= 629:
        return 'Genitourinary system'
    elif 630 <= num <= 679:
        return 'Pregnancy/Childbirth complications'
    elif 680 <= num <= 709:
        return 'Skin & subcutaneous tissue'
    elif 710 <= num <= 739:
        return 'Musculoskeletal & connective tissue'
    elif 740 <= num <= 759:
        return 'Congenital anomalies'
    elif 760 <= num <= 779:
        return 'Perinatal conditions'
    elif 780 <= num <= 799:
        return 'Symptoms & ill-defined conditions'
    elif 800 <= num <= 999:
        return 'Injury & poisoning'
    else:
        return 'Other'
```

```python
# First, define your groups (as you provided)
high_frequency = ['InternalMedicine', 'Family/GeneralPractice', 'Cardiology', 'Surgery-General',
                  'Orthopedics', 'Orthopedics-Reconstructive', 'Emergency/Trauma', 'Urology',
                  'ObstetricsandGynecology','Psychiatry','Pulmonology ','Nephrology','Radiologist']

low_frequency = ['Surgery-PlasticwithinHeadandNeck','Psychiatry-Addictive','Proctology','Dermatology',
                 'SportsMedicine','Speech','Perinatology','Neurophysiology','Resident','Pediatrics-Hematology-Oncology',
                 'Pediatrics-EmergencyMedicine','Dentistry','DCPTEAM','Psychiatry-Child/Adolescent',
                 'Pediatrics-Pulmonology','Surgery-Pediatric','AllergyandImmunology','Pediatrics-Neurology',
                 'Anesthesiology','Pathology','Cardiology-Pediatric','Endocrinology-Metabolism','PhysicianNotFound',
                 'Surgery-Colon&Rectal','OutreachServices','Surgery-Maxillofacial','Rheumatology','Anesthesiology-Pediatric',
                 'Obstetrics','Obsterics&Gynecology-GynecologicOnco']

pediatrics = ['Pediatrics','Pediatrics-CriticalCare','Pediatrics-EmergencyMedicine','Pediatrics-Endocrinology',
              'Pediatrics-Hematology-Oncology','Pediatrics-Neurology','Pediatrics-Pulmonology',
              'Anesthesiology-Pediatric','Cardiology-Pediatric','Surgery-Pediatric']

psychic = ['Psychiatry-Addictive', 'Psychology', 'Psychiatry',  'Psychiatry-Child/Adolescent',
           'PhysicalMedicineandRehabilitation', 'Osteopath']

neurology = ['Neurology', 'Surgery-Neuro',  'Pediatrics-Neurology', 'Neurophysiology']

surgery = ['Surgeon', 'Surgery-Cardiovascular','Surgery-Cardiovascular/Thoracic','Surgery-Colon&Rectal',
           'Surgery-General','Surgery-Maxillofacial','Surgery-Plastic','Surgery-PlasticwithinHeadandNeck',
           'Surgery-Thoracic','Surgery-Vascular','SurgicalSpecialty','Podiatry']

ungrouped = ['Endocrinology','Gastroenterology','Gynecology','Hematology','Hematology/Oncology','Hospitalist',
             'InfectiousDiseases','Oncology','Ophthalmology','Otolaryngology','Pulmonology','Radiology']

missing = ['Unknown']

def map_medical_specialty_to_num(spec):
    if spec in high_frequency:
        return 1
    elif spec in low_frequency:
        return 2
    elif spec in pediatrics:
        return 3
    elif spec in psychic:
        return 4
    elif spec in neurology:
        return 5
    elif spec in surgery:
        return 6
    elif spec in ungrouped:
        return 7
    elif spec in missing or pd.isna(spec):
        return 8
    else:
        return 9  # Catch-all for any unexpected specialty

df['medical_specialty_group_num'] = df['medical_specialty'].apply(map_medical_specialty_to_num)

```

```python
# Calculate percentage of rows that have the same value per column
constant_percentage = {}

for col in df.columns:
    most_freq = df[col].value_counts().max()
    total = df.shape[0]
    constant_percentage[col] = (most_freq / total) * 100

# Convert to a DataFrame for easy viewing
constant_df = pd.DataFrame.from_dict(constant_percentage, orient='index', columns=['most_common_value_percentage'])
constant_df = constant_df.sort_values(by='most_common_value_percentage', ascending=False)

print(constant_df)
```

```python
# Calculate % of most frequent value per column
constant_percentage = df.apply(lambda x: x.value_counts(normalize=True).max() * 100)

# Columns to drop: ≥ 99% constant
cols_to_drop = constant_percentage[constant_percentage >= 99].index.tolist()

# Drop those columns
df.drop(columns=cols_to_drop, inplace=True)

print(f"Dropped {len(cols_to_drop)} columns with ≥ 99% constant values.")
print("Remaining columns:", df.shape[1])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


race_map = {
    'Caucasian': 1,
    'AfricanAmerican': 2,
    'Hispanic': 3,
    'Asian': 4,
    'Other': 5,
    'Unknown': 6
}
df['race_num'] = df['race'].map(race_map)
df['gender_num'] = df['gender'].map({'Male':0, 'Female':1,'Unknown/Invalid': -1 })


age_color = "#A78BFA"        # light purple
race_colors = ["#D8B4FE", "#C084FC", "#A855F7", "#9333EA", "#7E22CE", "#6B21A8"]
gender_colors = ["#C084FC", "#9333EA", "#6B21A8"]

# -----------------------
# 1. Age histogram (full width)
# -----------------------
plt.figure(figsize=(10,5))
sns.histplot(df['age'], bins=10, kde=True, color=age_color)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# -----------------------
# 2. Pie charts for race and gender side by side
# -----------------------
fig, axes = plt.subplots(1, 2, figsize=(12,6))

# Race pie chart
race_counts = df['race'].value_counts()
axes[0].pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', colors=race_colors, startangle=140, textprops={'fontsize':12})
axes[0].set_title("Race Distribution", fontsize=14)

# Gender pie chart
gender_counts = df['gender'].value_counts()
axes[1].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=gender_colors, startangle=140, textprops={'fontsize':12})
axes[1].set_title("Gender Distribution", fontsize=14)

plt.tight_layout()
plt.show()
```

Highest number of encounters are between **60-80 in age**.

Most encounters with **Caucasian** while least from **Asian**

Number of encounters in **Female** is higher than **Male**

### Admission & Hospital Stay Details

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# List of numerical and categorical features
categorical_features = ['admission_type_id', 'discharge_disposition_id', 'admission_source_id', 'medical_specialty_group_num']
numerical_features = ['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications',
                      'number_outpatient', 'number_emergency', 'number_inpatient']

# -----------------------
# 1. Categorical features - count plots
# -----------------------
fig, axes = plt.subplots(2, 2, figsize=(16,10))

for ax, col in zip(axes.flatten(), categorical_features):
    sns.countplot(data=df, x=col, ax=ax, palette="Purples")
    ax.set_title(f"{col} Distribution")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")

plt.tight_layout()
plt.show()

# -----------------------
# 2. Numerical features - histograms + boxplots
# -----------------------
fig, axes = plt.subplots(len(numerical_features), 2, figsize=(14, 5*len(numerical_features)))

for i, col in enumerate(numerical_features):
    # Histogram
    sns.histplot(df[col], bins=20, kde=True, color="#A78BFA", ax=axes[i,0])
    axes[i,0].set_title(f"{col} Distribution (Histogram)")
    axes[i,0].set_xlabel(col)
    axes[i,0].set_ylabel("Count")

    # Boxplot
    sns.boxplot(data=df, y=col, color="#9333EA", ax=axes[i,1])
    axes[i,1].set_title(f"{col} Distribution (Boxplot)")

plt.tight_layout()
plt.show()
```

```python
categorical_features = [
    'max_glu_serum',
    'A1Cresult',
    'diag_1_group',
    'diag_2_group',
    'diag_3_group'
]

fig, axes = plt.subplots(3, 2, figsize=(16,14))
axes = axes.flatten()

for i, col in enumerate(categorical_features):
    ax = axes[i]
    sns.countplot(data=df, x=col, palette="Purples", ax=ax)
    ax.set_title(f"{col} Distribution")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=45)

    # Add percentages on top of bars
    total = len(df)
    for p in ax.patches:
        height = p.get_height()
        pct = height / total * 100
        ax.text(
            p.get_x() + p.get_width() / 2.,  # X position: middle of bar
            height + 5,                      # Y position: a little above bar
            f'{pct:.1f}%',                   # Text
            ha="center", va="bottom", fontsize=10
        )

# Remove empty subplot if exists
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

### Drugs

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

med_features = [
    'metformin', 'repaglinide', 'glimepiride', 'glipizide',
    'glyburide', 'pioglitazone', 'rosiglitazone',
    'insulin', 'diabetesMed', 'change'
]

fig, axes = plt.subplots(5, 2, figsize=(16,18))
axes = axes.flatten()

for i, col in enumerate(med_features):
    ax = axes[i]
    sns.countplot(data=df, x=col, ax=ax, palette="Purples")
    ax.set_title(f"{col} Distribution")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=30)

    # Add percentages on top of bars
    total = len(df)
    for p in ax.patches:
        height = p.get_height()
        pct = height / total * 100
        ax.text(
            p.get_x() + p.get_width() / 2.,  # x-position: center of bar
            height + 5,                      # y-position: slightly above bar
            f'{pct:.1f}%',                   # text to display
            ha='center', va='bottom', fontsize=10
        )

plt.tight_layout()
plt.show()
```
