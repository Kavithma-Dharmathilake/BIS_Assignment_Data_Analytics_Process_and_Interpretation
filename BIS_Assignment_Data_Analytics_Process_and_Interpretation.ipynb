### Initialize a new Git repository

To begin using Git for version control, you need to initialize a new Git repository in your project directory. Run the following command:

```bash
!git init
```

After initializing the repository, you can then configure your user name and email (if you haven't already), and proceed with staging and committing your files.

### Step 2: Add the Remote Repository URL

Now that you have created an empty repository on GitHub, you need to connect your local Git repository to this new remote one.

1.  **Locate the Remote URL:** On your newly created GitHub repository page, look for the green 'Code' button. Click on it, and you'll see options for cloning the repository (HTTPS, SSH, GitHub CLI). Copy the **HTTPS** URL.
    *Example URL:* `https://github.com/your-username/your-repository-name.git`
2.  **Open your Local Project Directory's Terminal:** Navigate to the root directory of your local project (the one that contains your `.git` folder) using your terminal or command prompt.
3.  **Add the Remote:** In your terminal, run the following command, replacing `<YOUR_GITHUB_REPOSITORY_URL>` with the URL you just copied:

    ```bash
!git remote add origin <YOUR_GITHUB_REPOSITORY_URL>
    ```

    *Explanation:* This command tells Git that your local `origin` (a common alias for the primary remote repository) now points to your GitHub repository.
4.  **Verify the Remote:** To confirm that the remote was added successfully, run:

    ```bash
!git remote -v
    ```

    You should see output similar to this, showing the fetch and push URLs for `origin`:

    ```
origin  https://github.com/your-username/your-repository-name.git (fetch)
origin  https://github.com/your-username/your-repository-name.git (push)
    ```
